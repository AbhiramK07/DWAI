```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model = "gemini-2.0-flash-exp")
```

```python
message = [("system", "You are a translator that translates English to Kannada. Translate the user sentence."), ("user", "I am Adarsh")]

```

```python
resp = llm.invoke(message)
resp
```

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
prompt = ChatPromptTemplate.from_messages([("system", "You are a translator that translates {input_lang} to {output_lang}. Translate the user sentence."), ("user", "{input_text}")])
```

```python
chain = prompt | llm
```

```python
chain.invoke({"input_lang" : "English", "output_lang" : "Telugu",
              "input_text" : "I am Abhiram"})
```
